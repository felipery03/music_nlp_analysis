```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/final_data.csv')
```

```python
df = df.drop(['lyrics', 'translated', 'ns_pt', 'orengo', 'no_stopwords', 'ns_pt'], axis =1)
```

```python
df.head()
```

```python
df['taxa_palavras'] = df['unique']/df['words']
```
