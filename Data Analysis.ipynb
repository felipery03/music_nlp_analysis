```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/final_data.csv')
```

```python
df = df.drop(['lyrics', 'translated', 'ns_pt', 'orengo', 'no_stopwords', 'ns_pt'], axis =1)
```

```python
df.head()
```

```python
df['avg_vocab'] = df[['unique', 'orengo_unique', 'unique_pt', 'orengo_pt']].mean(axis=1)
```

```python
df['avg_vocab_ns'] = df[['ns_unique', 'ns_orengo_unique', 'ns_uni_pt', 'ns_or_pt']].mean(axis=1)
```

```python
df['avg_song_vocab'] = df['avg_vocab']/25
```

```python
df['avg_song_vocab_ns'] = df['avg_vocab_ns']/25
```

```python
df['taxa_trad'] = df['foreign']/df['words']
```

```python
df.head()
```

```python
df.describe()
```
