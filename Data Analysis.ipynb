```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/final_data.csv')
```

```python
df.head()
```

```python
k = df.lyrics.str.cat(sep=' ')
n = len(set(k.split(' ')))
print(n)
```

```python
df = df.drop(['lyrics', 'orengo', 'no_stopwords'], axis =1)
```

```python
df['avg_vocab'] = df[['unique', 'orengo_unique', 'ns_unique', 'ns_orengo_unique']].mean(axis=1)
```

```python
df['avg_song_vocab'] = df['avg_vocab']/25
```

```python
df.head()
```

```python
df.describe()
```

## TOP 10 - Maiores vocabulários médios

```python
top_avg_vocab = df.sort_values('avg_vocab', ascending=False).nlargest(10, 'avg_vocab')
top_avg_vocab = top_avg_vocab.drop(['orengo_unique', 'ns_unique', 'ns_orengo_unique', 'avg_song_vocab', 'unique'], axis=1)
top_avg_vocab
```

## TOP 10 - Menores vocabulários médios

```python
bottom_avg_vocab = df.sort_values('avg_vocab', ascending=False).nsmallest(10, 'avg_vocab')
bottom_avg_vocab = bottom_avg_vocab.drop(['orengo_unique', 'ns_unique', 'ns_orengo_unique', 'avg_song_vocab', 'unique'], axis=1)
bottom_avg_vocab
```

## TOP 10 - Maiores vocabulários médios por música

```python
top_avg_song_vocab = df.sort_values('avg_song_vocab', ascending=False).nlargest(10, 'avg_song_vocab')
top_avg_song_vocab = top_avg_song_vocab.drop(['orengo_unique', 'ns_unique', 'ns_orengo_unique', 'avg_vocab', 'unique'], axis=1)
top_avg_song_vocab
```

## TOP 10 - Menores vocabulários médios por música

```python
bottom_avg_song_vocab = df.sort_values('avg_song_vocab', ascending=False).nsmallest(10, 'avg_song_vocab')
bottom_avg_song_vocab = bottom_avg_song_vocab.drop(['orengo_unique', 'ns_unique', 'ns_orengo_unique', 'avg_vocab', 'unique'], axis=1)
bottom_avg_song_vocab
```
